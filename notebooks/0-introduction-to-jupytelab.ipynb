Jupyter notebooks also allow us to execute external commands, right from the notebooks.

To execute an external command, just add a `!` before the command that you want to execute. Jupyter notebook understands that and treats whatever you have entered as external command and invokes it instread of execute it as code.

## Creating Files

Sometimes you may want to create a new file right from the jupyter notebook. 

To create a new file, start the cell with a line `%% what-ever-filename` and every thing after that cell gets saved int to given filename.

The following cell creates a file `five.txt` and the next cell counts the number of lines in it using the unix command `wc`.


```python
%%file five.txt
one
two
three
four
five
```

It is often handy to create python scripts right in the jupyter notebook and execute them. For example, the following cell creates a file hello.py and the next cell executes its using `python`.

```python
%%file hello.py

print("hello world")
```

```python
!python hello.py
```
